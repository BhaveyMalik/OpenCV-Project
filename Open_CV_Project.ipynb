# ***Converting any image into Pencil Sketch and Watercolor!***

# Importing required libraries

```python
import numpy as np 
import matplotlib.pyplot as plt
import cv2 
```

# Importing an image 

```python
img = cv2.imread("/content/R.jfif") 
plt.imshow(img)
```

# Converting image into RGB  

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

# Converting Image To Grayscale

```python
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap = 'gray')
```

# Smoothening Image

```python
blur = cv2.GaussianBlur(gray,(5,5),2)
plt.imshow(blur, cmap = 'gray')
```

```python
img_inv=cv2.bitwise_not(blur)
plt.imshow(img_inv)
```

# Detecting the Edges

```python
sobelx = cv2.Sobel(blur,cv2.CV_64F,1,0,ksize=5) 
sobely = cv2.Sobel(blur,cv2.CV_64F,0,1,ksize=5) 
gradmag_sq = np.square(sobelx)+np.square(sobely)  
gradmag = np.sqrt(gradmag_sq) 
plt.imshow(gradmag, cmap ='gray')
```

# Inverting the Image

```python
gradmag_inv = 255-gradmag 
plt.imshow(gradmag_inv, cmap = "gray")
```

# Getting the Threshold of the Image

```python
thresh_value, thresh_img = cv2.threshold(gradmag_inv,10,255,cv2.THRESH_BINARY)
plt.imshow(thresh_img, cmap = 'gray')
```

# Pencil Sketch Effect

```python
pencilsketch_gray, pencilsketch_color  = cv2.pencilSketch(img_rgb, sigma_s=60, sigma_r=0.07, shade_factor=0.05) 

plt.imshow(pencilsketch_gray, cmap ='gray') 
```

# Coloured Pencil Sketch Effect 

```python
plt.imshow(pencilsketch_color) 
```

# Water Colour Effect

```python
waterpainting = cv2.stylization(img_rgb, sigma_s=3, sigma_r=3)
plt.imshow(waterpainting)
```
